```python
from statsmodels.tsa.stattools import adfuller

#perform augmented Dickey-Fuller test
p_value=adfuller(berkshire_data['Open'])[1]
print(f'The p value is {p_value}')
```

```python
berkshire_data['open_first_difference']=berkshire_data['Open']-berkshire_data['Open'].shift(1)# We took a  difference
```

```python
berkshire_data['open_seasonal_difference']=berkshire_data['Open']-berkshire_data['Open'].shift(12)# We took a  difference
```

```python
berkshire_data['open_second_difference']=berkshire_data['open_first_difference']-berkshire_data['open_first_difference'].shift(1)# We took a  difference
```

```python
from statsmodels.tsa.stattools import adfuller

p_value = adfuller(berkshire_data['open_first_difference'].dropna())[1]
print(f'The p value is {p_value}')# p-value 0 makes it a stationary data 
```

```python
from statsmodels.tsa.stattools import adfuller

p_value = adfuller(berkshire_data['open_second_difference'].dropna())[1]
print(f'The p value is {p_value}')# p-value 0 makes it a stationary data 
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf 
```

```python
acf=plot_acf(berkshire_data['open_second_difference'].dropna())# q value should be 1 12 or 16
```

```python
pacf=plot_pacf(berkshire_data['open_second_difference'].dropna())# p value should be 1 or 2 or 3 |
```

```python
acf=plot_acf(berkshire_data['open_seasonal_difference'].dropna())# q value should be 1 12 or 16
```

```python
pacf=plot_pacf(berkshire_data['open_seasonal_difference'].dropna())# p value should be 1 or 2 or 3 |
```

```python
from datetime import datetime,timedelta
train_data_end=datetime(2022,12,31)
test_data_end=datetime(2023,9,5)
```

```python
train_data=berkshire_data[:train_data_end]
test_data=berkshire_data[train_data_end:test_data_end]
train_data
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
train_data
```

```python
model_ARIMA=ARIMA(berkshire_data['Open'],order=(1,2,1))
```

```python
model_arima_fit=model_ARIMA.fit()
model_arima_fit.summary()
```

```python
pred_start_date=test_data.index[0]
pred_end_date=test_data.index[-1]
```

```python
test_data.head(234)
```

```python
pred=model_arima_fit.predict(start=pred_start_date,end=pred_end_date)
```

```python
residuals=test_data['Open']-pred
```

```python
residuals
```

```python
model_arima_fit.resid.plot(kind='kde')
```

```python
test_data['Predicted_ARIMA']=pred
```

```python
test_data[['Open','Predicted_ARIMA']].plot()
```

```python
model_SARIMA = SARIMAX(berkshire_data['Open'], order=(1,0,1), seasonal_order=(0,1,0,12))
# we took d to be 0 in the first as the open seasonal differencing will be covered in seasonal order 
```

```python
results_SARIMA = model_SARIMA.fit()
```

```python
pred1=results_SARIMA.predict(start=pred_start_date,end=pred_end_date)
```

```python
residuals=test_data['Open']-pred1
```

```python
residuals.plot(kind='kde')
```

```python
test_data['Predicted_SARIMA']=pred1
```

```python
test_data[['Open','Predicted_SARIMA']].plot()
```
